```python
from keras.models import Model
```

# Single Input, Multiple Output

```python
model = Model(inputs= x, outputs=[output1,output2])
```

```python
from keras.layers import *
```

```python
x = Input(shape=(3,))
```

```python
hidden1 = Dense(128, activation='relu')(x)
hidden2 = Dense(64, activation='relu')(hidden1)

output1 = Dense(1, activation='relu')(hidden2)
output2= Dense(1, activation='sigmoid')(hidden2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```

# Multiple Input, Single Output

```python
inputA = Input(shape=(32,))
inputB = Input(shape=(128,))
```

```python
x1 = Dense(8, activation='relu')(inputA)
x2 = Dense(4, activation='relu')(x1)
```

```python
y1 = Dense(64, activation='relu')(inputB)
y2 = Dense(32, activation='relu')(y1)
y3 = Dense(4,activation='relu')(y2)
```

```python
combined = concatenate([x2,y3])
```

```python
z1 = Dense(2,activation='relu')(combined)
z2 = Dense(1,activation='linear')(z1)
```

```python
model1 = Model(inputs=[inputA,inputB],outputs=z2)
```

```python
plot_model(model1,show_shapes=True)
```
